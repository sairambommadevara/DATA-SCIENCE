```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
x=np.linspace(0,10,200)
#print(x)
y=x**2
plt.plot(x,y)
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('plot of  y=x**2')
plt.show()
```

```
plt.subplot(1,2,1)
plt.plot(x,y,'r*-')
plt.subplot(1,2,2)
plt.plot(y,x,'g--')
plt.show()
```

# **SEABORN**

```
import seaborn as sns
```

```
tips=sns.load_dataset('tips')
```

```
tips.head()
```

## distribution plot

```
sns.distplot(tips['tip'])
```

joint plot

```
sns.jointplot(x=tips['total_bill'],y=tips['tip'])
```

```
sns.jointplot(x=tips['total_bill'],y=tips['tip'],kind='hex')
```

```
sns.jointplot(x=tips['total_bill'],y=tips['tip'],kind='scatter')
```

```
sns.jointplot(x=tips['total_bill'],y=tips['tip'],kind='reg')
```

```
sns.jointplot(x=tips['total_bill'],y=tips['tip'],kind='resid')
```

```
sns.jointplot(x=tips['total_bill'],y=tips['tip'],kind='kde')
```

### paire plot

```
sns.pairplot(tips)
```

```
sns.pairplot(tips,hue='sex')
```

bar plot

```
sns.barplot(x='sex',y='tip',data=tips)
```

```
sns.countplot(x='sex',data=tips)
```

```
sns.countplot(x='day',data=tips)
```
